#### 2.3 Is Data is ROCCC or not?
Good data is ROCCC: Reliable, Orginal, Comprehensive, Current, and Cited
* **Reliable: No** - Data only has 33 respondents so it is unreliable as survey have not all responses.
* **Original: No** - Data was provided by Amazon Mechanical Turk so it is not original
* **Comprehensive: Yes** - FitBit Data aligns with the majority of Bellabeat products
* **Current: No** - Data is 6 years old so it is possible that it has anomalies
* **Cited: No** - Data was collected from at third party (Amazon Mecanical Turk)

#### Transforming data
* Change _ActivityDate_ to datetime data type
* Convert the format of _ActivityDate_ to yyyy-mm-dd
* Create Weekday column for easier analysis
* Create a _TotalMinutes_ column by adding the Acitvity Minutes columns together
* Rename columns

```python
# Change ActivityDate to datetime64 data type
daily_activity["ActivityDate"]  = pd.to_datetime(daily_activity["ActivityDate"], format = "%m/%d/%Y")
daily_activity.sample(5).iloc[:,-3:]

# Look at information
daily_activity.info()

# Get an detail overview of the data
daily_activity.head()
```

```python
# Renaming of column daily_activity to di_activity
new_cols = ['Id', 'ActivityDate', 'WeekDay','TotalSteps', 'TotalDistance',
                'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance','SedentaryActiveDistance','VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes','SedentaryMinutes','Calories']

di_activity = daily_activity.reindex(columns=new_cols)

# Make sure the new column was added
di_activity.head()
```

#### Create _WeekDay_ column for easier analysis

```python
# Add a WeekDay column 
di_activity["WeekDay"] = di_activity["ActivityDate"].dt.day_name()

# Check to make sure the week days were added
# head(7) means seven days will print
di_activity["WeekDay"].head(7)
```

#### Create _TotalMinutes_ column

```python
# Create new column TotalMinutes by adding activity minutes
di_activity["TotalMinutes"] = di_activity["VeryActiveMinutes"] + di_activity["FairlyActiveMinutes"] + di_activity["LightlyActiveMinutes"] + di_activity["SedentaryMinutes"]
di_activity["TotalMinutes"].head()
```

#### Create _TotalHours_ column for easier analysis

```python
# Create new column TotalHours by converting TotalMinutes column
di_activity["TotalHours"] = round(di_activity["TotalMinutes"] / 60)
di_activity["TotalHours"].head()
```

```python
di_activity["TotalSeconds"] = round(di_activity["TotalMinutes"] * 60)
di_activity["TotalSeconds"].head()
```

```python
# Looking at Calories Burned vs Total Steps Taken

#plotting the scatterplot
plt.figure(figsize=(9,7)) # specify size of the chart
plt.scatter(di_activity.TotalSteps, di_activity.Calories, c = di_activity.Calories, 
            cmap = "inferno")

# Adding the average, min and max amount of steps and calories
avg_calories = 2303
avg_steps = 7637
min_calories = 0
min_steps = 0
max_calories = 4900
max_steps = 36019

plt.colorbar(orientation = "vertical")
plt.axvline(avg_steps, color = "Green", label = "Average steps")
plt.axhline(avg_calories, color = "violet", label = "Average calories burned")
plt.xlabel("Amount of Steps taken")
plt.ylabel("Amount of Calories burned")
plt.title("calories burned vs steps taken")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Looking at the Frequency of FitBit usage 

#plotting the histogram
plt.figure(figsize=(7,6)) 
plt.hist(di_activity.WeekDay, bins = 7, 
         width = 0.5, color = "red", edgecolor = "black")

# adding labels
plt.xlabel("Days of the week")
plt.ylabel("Number of usage time")
plt.title("No of times any user used fitbit app during the week")
plt.show()
```
